# Definicion de neurona

Unidad basica de procesamiento de una red neuronal, tiene conexiones de entrada en la cual reciben estimulos y realiza una suma ponderada por el peso que se asigna por cada una de las entradas



La reurona internamente realiza es una regresion lineal o similar
