Predecir un resultado a partir de datos de entrada con el cual se puede obtener una solucion

Algoritmos:
Clustering, arboles de decision, series temporales, Bayes, procesamiento de lenguaje natural.
Contraste de hipotesis, regresion lineal, regresion logistica
    <b>Supervisados: </b>
    <b>No supervisados: </b>

Herramientas de estadistica: R y Python.

/////////////////////////////////////////////

El modelo se construye a partir de datos pasados
Averiguan los datos que faltan
Se dividen los datos en dos conjuntos:
    1. Entrenamiento
    2. Validacion
#Nota: por lo general se parte los datos en 80-20
 
El modelo termina como una funcion matematica como una regresion.
