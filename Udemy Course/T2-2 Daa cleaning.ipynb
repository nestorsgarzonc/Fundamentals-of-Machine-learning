# Simulacion de monte carlo

* generar dos numeros aleatorios entre 0 y 1 (x,y)
*  Calcular $x^2$, $y^2$
    * Si el valor es <1 esta dentro del circulo
    * Si el valor es >1 esta fuera del circulo
* Calcular numero de veces que esta dentro del circulo y se divide entre el numero total de intentos para obtener una aproximacion de caer en el circulo
* Usar probabiidad para calcular pi
* Repetir experimento para obtener diferentes aproximaciones de pi (aprox 1000 veces)
* Calcular promedio 1000 experimentos para dar un valor final
* Multiplicar por 4 y ser feliz

```python
def pi_montecarlo():
    a=0
    b=1
    n=1000
    pi=0
    pi_list=[]
    for j in range(n):
        dentro=0
        x=np.random.uniform(a,b,n).tolist()
        y=np.random.uniform(a,b,n).tolist()

        for i in range(n):
            x[i]=x[i]*x[i]
            y[i]=y[i]*y[i]

        for i in range(1000):
            if(x[i]+y[i]<=1):
                dentro+=1

        pi_prom=dentro/(1000)*4
        pi_list.append(pi_prom)
        pi+=pi_prom

    pi=pi/n
    print(pi)
    fig=plt.hist(pi_list, 20)
    plt.title('PI number')
    return pi, fig
```

```python
x, fig=pi_montecarlo()
```
